```python
import pyautogui
import time
import pygetwindow as gw

pyautogui.FAILSAFE = True

MESSAGE = "hello"
CHAT_INTERVAL = 10
SCAN_INTERVAL = 3
CONFIDENCE = 0.65   

TRADE_COOLDOWN = 5

TRADE_IMAGE = "C:\\Users\\konra\\Desktop\\trade.png"
ACCEPT1_IMAGE = "C:\\Users\\konra\\Desktop\\accept.png"
ACCEPT2_IMAGE = "C:\\Users\\konra\\Desktop\\accept2.png"

WINDOW_TITLE = "RuneLite"

print("â³ Start za 3 sekundy â€“ kliknij w okno RuneLite")
time.sleep(3)

last_chat = 0
last_trade_click = 0

def get_runelite_window():
    wins = gw.getWindowsWithTitle(WINDOW_TITLE)
    return wins[0] if wins else None

while True:
    now = time.time()

    # ðŸ‘‰ TYLKO AKTYWACJA OKNA
    win = get_runelite_window()
    if win and not win.isActive:
        win.activate()
        time.sleep(5)

    # 1. Chat
    if now - last_chat >= CHAT_INTERVAL:
        pyautogui.typewrite(MESSAGE, interval=0.05)
        pyautogui.press("enter")
        last_chat = now

    # 2. Czy handel otwarty
    trade_open = False
    for img in (ACCEPT1_IMAGE, ACCEPT2_IMAGE):
        try:
            if pyautogui.locateOnScreen(img, confidence=CONFIDENCE):
                trade_open = True
        except:
            pass

    # 3. Trade
    if not trade_open and now - last_trade_click >= TRADE_COOLDOWN:
        try:
            trade = pyautogui.locateOnScreen(TRADE_IMAGE, confidence=CONFIDENCE)
            if trade:
                x, y = pyautogui.center(trade)
                pyautogui.moveTo(x, y, duration=0.1)
                pyautogui.click()
                last_trade_click = now
                print("ðŸ’¬ KlikniÄ™to TRADE")
        except:
            pass

    # 4. Accepty
    for img in (ACCEPT1_IMAGE, ACCEPT2_IMAGE):
        try:
            btn = pyautogui.locateOnScreen(img, confidence=CONFIDENCE)
            if btn:
                x, y = pyautogui.center(btn)
                pyautogui.moveTo(x, y, duration=0.1)
                pyautogui.click()
                print("âœ… KlikniÄ™to ACCEPT")
        except:
            pass

    time.sleep(SCAN_INTERVAL)
```
